What is a missing value? Some columns have ? instead of real data (this dataset uses ? as its missing value marker, not the usual NaN). We need to fix this.

```python
df.replace("?", np.nan, inplace=True)

missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct.round(1)
})
print(missing_df[missing_df['missing_count'] > 0])
```

```python
# Drop columns — too many missing values to be useful
cols_to_drop = [
    'weight',           # 96.9% missing
    'max_glu_serum',    # 94.7% missing
    'A1Cresult',        # 83.3% missing
    'medical_specialty',# 49.1% missing
    'payer_code'        # 39.6% missing
]
df.drop(columns=cols_to_drop, inplace=True)

# Drop rows — only a tiny % missing, safe to remove
df.dropna(subset=['race', 'diag_1', 'diag_2', 'diag_3'], inplace=True)

# Confirm everything is clean now
print("Rows remaining:", len(df))
print("Columns remaining:", df.shape[1])
print("Missing values left:", df.isnull().sum().sum())
```

What is the target column? The column your model is trying to predict. Here it is readmitted. Currently it has 3 values: <30, >30, NO. We convert it to binary — 1 means high risk (readmitted in under 30 days), 0 means not high risk.

```python
df['readmitted_binary'] = (df['readmitted'] == '<30').astype(int)

print("Value counts:")
print(df['readmitted_binary'].value_counts())
print()
print("High risk patients:", df['readmitted_binary'].sum())
print("% high risk:", round(df['readmitted_binary'].mean() * 100, 1), "%")
```

Why? Age is stored as text ranges like [50-60). The model cannot do maths with text. We convert each range to the midpoint number.

```python
age_map = {
    '[0-10)':   5,
    '[10-20)':  15,
    '[20-30)':  25,
    '[30-40)':  35,
    '[40-50)':  45,
    '[50-60)':  55,
    '[60-70)':  65,
    '[70-80)':  75,
    '[80-90)':  85,
    '[90-100)': 95
}

df['age'] = df['age'].map(age_map)

print("Age column converted!")
print(df['age'].value_counts().sort_index())
```

Expected output — you will see numbers like 45, 55, 65, 75 with their counts. The most common age group will be 70-80 year olds (seniors are readmitted most often).

Why? Gender is stored as Male / Female text. We convert to 1 and 0 so the model can use it.

```python
df['gender'] = df['gender'].map({
    'Male':             1,
    'Female':           0,
    'Unknown/Invalid':  0
})

print("Gender encoded!")
print(df['gender'].value_counts())
```

Why? Race has 5 categories (Caucasian, AfricanAmerican, Hispanic, Asian, Other). We cannot just use numbers 1-5 here because that would imply "Asian = 4 times more than Caucasian" which makes no sense. Instead we use get_dummies which creates a separate 0/1 column for each category.

Before (1 column)
race = "Caucasian"
race = "AfricanAmerican"
race = "Hispanic"
→
After (4 columns)
race_Hispanic=0, race_Other=0...
race_AfricanAmerican=1, ...
race_Hispanic=1, ...

```python
df = pd.get_dummies(df, columns=['race'], drop_first=True)

print("Race encoded into columns!")
race_cols = [c for c in df.columns if c.startswith('race_')]
print("New race columns:", race_cols)
print("Shape now:", df.shape)
```

The drop_first=True drops one of the race columns on purpose. This avoids a maths problem called "dummy variable trap" — you do not need to understand it deeply now, just always use drop_first=True for get_dummies.
Expected output — you will see 4 new columns like race_Asian, race_Caucasian, race_Hispanic, race_Other. Shape will now be 49 columns (went up because race became 4 columns).

```python
print("=== STEP 4 COMPLETE — FINAL CHECK ===")
print()
print("Shape:", df.shape)
print()
print("Target column sample:")
print(df['readmitted_binary'].value_counts())
print()
print("Age column (should be numbers now):")
print(df['age'].unique())
print()
print("Gender column (should be 0 and 1 only):")
print(df['gender'].unique())
print()
print("Any missing values?", df.isnull().sum().sum())
print()
print("All good! Ready for Step 5.")
```

Feature engineering means creating new columns by combining or transforming existing ones. The raw dataset has columns like number_emergency and total_visits separately — but the ratio of those two tells a much stronger story about a patient's risk than either column alone.



What does the +1 do in emergency_rate? We add 1 to total_visits to avoid dividing by zero — if a patient somehow has 0 total visits, dividing by zero would crash your code. This is called "Laplace smoothing" — sound smart when you say it in interviews.

```python
df['total_visits'] = (
    df['number_outpatient'] +
    df['number_emergency'] +
    df['number_inpatient']
)

df['emergency_rate'] = (
    df['number_emergency'] /
    (df['total_visits'] + 1)
).round(3)

print("total_visits sample:")
print(df['total_visits'].describe().round(2))
print()
print("emergency_rate sample:")
print(df['emergency_rate'].describe().round(3))
```

What is this measuring? Number of medications divided by days in hospital. A patient on 20 medications for only 1 day is being discharged with a very complex drug regimen — high risk of not managing it properly at home.

Patient A — low density
num_medications = 5
time_in_hospital = 10
density = 5/11 = 0.45
Patient B — high density
num_medications = 18
time_in_hospital = 2
density = 18/3 = 6.0

```python
df['medication_density'] = (
    df['num_medications'] /
    (df['time_in_hospital'] + 1)
).round(3)

print("medication_density created!")
print(df['medication_density'].describe().round(3))
print()
print("Sample — high density patients (top 5):")
print(df[['num_medications','time_in_hospital',
          'medication_density']].nlargest(5, 'medication_density'))
```

diagnosis_count counts how many of the 3 diagnosis columns (diag_1, diag_2, diag_3) are filled in. A patient with 3 diagnoses is more complex than one with 1.

is_senior is a simple 1/0 flag — 1 if patient is 65 or older, 0 if younger. Senior patients have much higher readmission rates.

```python
df['diagnosis_count'] = (
    df[['diag_1', 'diag_2', 'diag_3']].notna().sum(axis=1)
)

df['is_senior'] = (df['age'] >= 65).astype(int)

print("diagnosis_count distribution:")
print(df['diagnosis_count'].value_counts().sort_index())
print()
print("Senior patients:", df['is_senior'].sum())
print("Non-senior patients:", (df['is_senior'] == 0).sum())
print()
print("Readmission rate — seniors:",
      round(df[df['is_senior']==1]['readmitted_binary'].mean()*100, 1), "%")
print("Readmission rate — non-seniors:",
      round(df[df['is_senior']==0]['readmitted_binary'].mean()*100, 1), "%")
```

```python
df['high_visit_risk'] = (
    (df['is_senior'] == 1) &
    (df['number_emergency'] >= 3)
).astype(int)

print("high_visit_risk flag created!")
print("High risk patients flagged:", df['high_visit_risk'].sum())
print()
print("Readmission rate in high_visit_risk group:",
      round(df[df['high_visit_risk']==1]['readmitted_binary'].mean()*100,1), "%")
print("Readmission rate in others:",
      round(df[df['high_visit_risk']==0]['readmitted_binary'].mean()*100,1), "%")
```

The high_visit_risk group will show a significantly higher readmission rate — possibly 18-25% vs the baseline 11%. This proves the interaction feature is capturing something real. This exact insight is what you explain in interviews.

Why drop these? Patient IDs are just tracking numbers — they have no clinical meaning. The original readmitted text column must be removed because we already converted it to readmitted_binary. Raw diagnosis codes (diag_1/2/3) are free text — we already captured their count in diagnosis_count.

```python
cols_to_remove = [
    'encounter_id',
    'patient_nbr',
    'readmitted'
]

existing = [c for c in cols_to_remove if c in df.columns]
df.drop(columns=existing, inplace=True)

if 'diag_1' in df.columns:
    df.drop(columns=['diag_1','diag_2','diag_3'], inplace=True)

print("Cleaned up!")
print("Final shape:", df.shape)
print()
print("Your 6 new features:")
new_feats = ['total_visits','emergency_rate',
             'medication_density','diagnosis_count',
             'is_senior','high_visit_risk']
print(df[new_feats].head(5).round(3))
```

```python
df.to_csv("../data/cleaned_data.csv", index=False)

print("=== PHASE 2 COMPLETE ===")
print()
print("Saved to: data/cleaned_data.csv")
print("Final rows:", len(df))
print("Final columns:", df.shape[1])
print("High risk patients:", df['readmitted_binary'].sum())
print("Missing values:", df.isnull().sum().sum())
print()
print("New features added:")
for f in ['total_visits','emergency_rate','medication_density',
          'diagnosis_count','is_senior','high_visit_risk']:
    print(f"  + {f}")
```
